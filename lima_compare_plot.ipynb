## MT-bench: lima comparison

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data_scale = '10k'
print(f"##### data scale: {data_scale} ############")

if data_scale == '1k':
    categories = ['Ours (M)-1k', 'Ours (G)-1k', 'Ours (L)-1k']

    win_rate_llama = np.array([36.3, 40.6, 35.0])
    tie_rate_llama = np.array([28.8, 28.1, 24.4])
    loss_rate_llama = np.array([35.0, 31.3, 40.6])

    win_rate_mistral = np.array([31.3, 32.5, 26.3])
    tie_rate_mistral = np.array([36.9, 40.0, 38.8])
    loss_rate_mistral = np.array([31.9, 27.5, 35.0])

elif data_scale == '10k':
    categories = ['Ours (M)-10k', 'Ours (G)-10k', 'Ours (L)-10k']
    win_rate_llama = np.array([46.3, 50.6, 51.9])
    loss_rate_llama = np.array([20.6, 21.9, 19.4])
    tie_rate_llama = np.array([33.1, 27.5, 28.8])

    win_rate_mistral = np.array([48.8, 47.5, 42.5])
    loss_rate_mistral = np.array([18.1, 17.5, 25.0])
    tie_rate_mistral = np.array([33.1, 35.0, 32.5])
else:
    print('data scale input error!')    

# Set bar width and y-axis positions
bar_width = 0.2
y_positions = np.arange(len(categories))

# Adjust spacing between bars
y_spacing = 0.4
y_positions = y_positions * y_spacing

# Adjust chart width
fig, axs = plt.subplots(1, 2, figsize=(9, 4), gridspec_kw={'wspace': 0})

# Subplot 1: LLaMA-3.1-8B Base Model
axs[0].barh(y_positions, win_rate_llama, color='#2D9F39', label='Win Rate', height=bar_width, alpha=0.8)
axs[0].barh(y_positions, tie_rate_llama, left=win_rate_llama, color='#F8C03F', label='Tie Rate', height=bar_width, alpha=0.8)
axs[0].barh(y_positions, loss_rate_llama, left=win_rate_llama + tie_rate_llama, color='#E87C2F', label='Loss Rate', height=bar_width, alpha=0.8)

for i in range(len(categories)):
    axs[0].text(win_rate_llama[i] / 2, y_positions[i], f'{win_rate_llama[i]}%', va='center', ha='center', color='black', fontsize=18)
    axs[0].text(win_rate_llama[i] + tie_rate_llama[i] / 2, y_positions[i], f'{tie_rate_llama[i]}%', va='center', ha='center', color='black', fontsize=18)
    axs[0].text(win_rate_llama[i] + tie_rate_llama[i] + loss_rate_llama[i] / 2, y_positions[i], f'{loss_rate_llama[i]}%', va='center', ha='center', color='black', fontsize=18)

# Set y-axis labels and positions
axs[0].set_yticks(y_positions)
axs[0].set_yticklabels(categories, fontsize=20)
axs[0].set_yticklabels([])

# Set spines and title
axs[0].spines['left'].set_edgecolor('gray')
axs[0].spines['left'].set_linewidth(1.0)
axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)
axs[0].spines['bottom'].set_visible(False)
axs[0].set_title('Base model: LLaMA-3.1-8B', pad=35, fontsize=20)
axs[0].set_xticks([])

# Subplot 2: Mistral-7B-v0.3 Base Model
axs[1].barh(y_positions, win_rate_mistral, color='#2D9F39', label='Win Rate', height=bar_width, alpha=0.8)
axs[1].barh(y_positions, tie_rate_mistral, left=win_rate_mistral, color='#F8C03F', label='Tie Rate', height=bar_width, alpha=0.8)
axs[1].barh(y_positions, loss_rate_mistral, left=win_rate_mistral + tie_rate_mistral, color='#E87C2F', label='Loss Rate', height=bar_width, alpha=0.8)

for i in range(len(categories)):
    axs[1].text(win_rate_mistral[i] / 2, y_positions[i], f'{win_rate_mistral[i]}%', va='center', ha='center', color='black', fontsize=18)
    axs[1].text(win_rate_mistral[i] + tie_rate_mistral[i] / 2, y_positions[i], f'{tie_rate_mistral[i]}%', va='center', ha='center', color='black', fontsize=18)
    axs[1].text(win_rate_mistral[i] + tie_rate_mistral[i] + loss_rate_mistral[i] / 2, y_positions[i], f'{loss_rate_mistral[i]}%', va='center', ha='center', color='black', fontsize=18)

# Set y-axis labels and positions
axs[1].set_yticks(y_positions)
axs[1].set_yticklabels([])

# Set legend and spines
axs[1].legend(['Win Rate', 'Tie Rate', 'Loss Rate'], loc='upper center', bbox_to_anchor=(0, 1.16), ncol=3, frameon=False, fontsize=18)
axs[1].spines['left'].set_edgecolor('gray')
axs[1].spines['left'].set_linewidth(1.0)
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)
axs[1].spines['bottom'].set_visible(False)
axs[1].set_title('Base model: Mistral-7B-v0.3', pad=35, fontsize=20)
axs[1].set_xticks([])

# Adjust layout
plt.tight_layout()

plt.show()

# lima_compare_mt_bench_1k_minipage
# lima_compare_mt_bench_10k_minipage
```
