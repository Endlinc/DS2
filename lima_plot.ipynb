```python
import matplotlib.pyplot as plt
import numpy as np

############ vicuna_bench ############

# categories = ['Ours(M)-1k', 'Ours(G)-1k', 'Ours(L)-1k']  # 顺序调整为从上到下显示 L, G, M
# categories = ['Ours(M)', 'Ours(G)', 'Ours(L)']  # 顺序调整为从上到下显示 L, G, M

data_scale = '10k' ##'10k'
print(f"##### data scale: {data_scale} ############")

if data_scale == '1k':
    categories = ['Ours(M)', 'Ours(G)', 'Ours(L)']  # 顺序调整为从上到下显示 L, G, M
    win_rate_llama = np.array([43.8, 47.5, 52.5])
    loss_rate_llama = np.array([43.8, 35.0, 27.5])
    tie_rate_llama = np.array([12.5, 17.5, 20.0])

    win_rate_mistral = np.array([42.5, 43.8, 40.0])
    loss_rate_mistral = np.array([37.5, 40.0, 41.3])
    tie_rate_mistral = np.array([20.0, 16.3, 18.8])
    
elif data_scale == '10k':
    # categories = ['Ours (M)-10k', 'Ours (G)-10k', 'Ours (L)-10k']  # 顺序调整为从上到下显示 L, G, M
    categories = ['Ours(M)', 'Ours(G)', 'Ours (L)']  # 顺序调整为从上到下显示 L, G, M

    win_rate_llama = np.array([45.0, 57.5, 47.5])

    loss_rate_llama = np.array([37.5, 30.0, 41.3])
    tie_rate_llama = np.array([17.5, 12.5, 11.3])

    win_rate_mistral = np.array([47.5, 57.5, 48.1])
    loss_rate_mistral = np.array([33.8, 30.0, 26.9])
    tie_rate_mistral = np.array([18.8, 12.5, 25.0])
else:
    print("data scale error!!")    



# 设置条形宽度
bar_width = 0.2  # 缩小条形的宽度
y_positions = np.arange(len(categories))  # 设置 y 轴位置

# 缩小条形之间的间距
y_spacing = 0.4  # 将每个条形的间隔设置得更紧凑
y_positions = y_positions * y_spacing

# 调整图表的宽度
fig, axs = plt.subplots(1, 2, figsize=(10, 4), gridspec_kw={'wspace': 0})

# 子图1: LLaMA-3.1-8B Base Model
axs[0].barh(y_positions, win_rate_llama, color='#2D9F39', label='Win Rate', height=bar_width, alpha=0.8)
axs[0].barh(y_positions, tie_rate_llama, left=win_rate_llama, color='#F8C03F', label='Tie Rate', height=bar_width, alpha=0.8)
axs[0].barh(y_positions, loss_rate_llama, left=win_rate_llama + tie_rate_llama, color='#E87C2F', label='Loss Rate', height=bar_width, alpha=0.8)

for i in range(len(categories)):
    axs[0].text(win_rate_llama[i] / 2, y_positions[i], f'{win_rate_llama[i]}%', va='center', ha='center', color='black',fontsize=17)
    axs[0].text(win_rate_llama[i] + tie_rate_llama[i] / 2, y_positions[i], f'{tie_rate_llama[i]}%', va='center', ha='center', color='black',fontsize=17)
    axs[0].text(win_rate_llama[i] + tie_rate_llama[i] + loss_rate_llama[i] / 2, y_positions[i], f'{loss_rate_llama[i]}%', va='center', ha='center', color='black',fontsize=17)

# 设置 y 轴的刻度标签和位置
axs[0].set_yticks(y_positions)
axs[0].set_yticklabels(categories, fontsize=20)

# 将 legend 放置在图表内部顶部
# axs[0].legend(loc='upper center', bbox_to_anchor=(0.5, 1.10), ncol=3, frameon=False,fontsize=11)

# 只保留左侧的竖线
axs[0].spines['left'].set_edgecolor('gray')  # 设置左侧边框颜色为黑色
axs[0].spines['left'].set_linewidth(1.0)          # 调整左侧边框线的宽度
axs[0].spines['top'].set_visible(False)           # 隐藏顶部边框
axs[0].spines['right'].set_visible(False)         # 隐藏右侧边框
axs[0].spines['bottom'].set_visible(False)        # 隐藏底部边框

# 图表设置
axs[0].set_title('Base model: LLaMA-3.1-8B', pad=35,fontsize=20)
axs[0].set_xticks([])

# 子图2: Mistral-7B-v0.3 Base Model
axs[1].barh(y_positions, win_rate_mistral, color='#2D9F39', label='Win Rate', height=bar_width, alpha=0.8)
axs[1].barh(y_positions, tie_rate_mistral, left=win_rate_mistral, color='#F8C03F', label='Tie Rate', height=bar_width, alpha=0.8)
axs[1].barh(y_positions, loss_rate_mistral, left=win_rate_mistral + tie_rate_mistral, color='#E87C2F', label='Loss Rate', height=bar_width, alpha=0.8)

for i in range(len(categories)):
    axs[1].text(win_rate_mistral[i] / 2, y_positions[i], f'{win_rate_mistral[i]}%', va='center', ha='center', color='black',fontsize=17)
    axs[1].text(win_rate_mistral[i] + tie_rate_mistral[i] / 2, y_positions[i], f'{tie_rate_mistral[i]}%', va='center', ha='center', color='black',fontsize=17)
    axs[1].text(win_rate_mistral[i] + tie_rate_mistral[i] + loss_rate_mistral[i] / 2, y_positions[i], f'{loss_rate_mistral[i]}%', va='center', ha='center', color='black',fontsize=17)

# 设置 y 轴的刻度标签和位置
axs[1].set_yticks(y_positions)
axs[1].set_yticklabels([])

# 将 legend 放置在图表内部顶部
# axs[1].legend(loc='upper center', bbox_to_anchor=(0.5, 1.10), ncol=3, frameon=False,fontsize=11)
axs[1].legend(['Win Rate', 'Tie Rate', 'Loss Rate'], loc='upper center', bbox_to_anchor=(0, 1.16), ncol=3, frameon=False, fontsize=18)

# 只保留左侧的竖线
axs[1].spines['left'].set_edgecolor('gray')  # 设置左侧边框颜色为黑色
axs[1].spines['left'].set_linewidth(1.0)          # 调整左侧边框线的宽度
axs[1].spines['top'].set_visible(False)           # 隐藏顶部边框
axs[1].spines['right'].set_visible(False)         # 隐藏右侧边框
axs[1].spines['bottom'].set_visible(False)        # 隐藏底部边框

# 图表设置
axs[1].set_title('Base model: Mistral-7B-v0.3', pad=35, fontsize=20)
axs[1].set_xticks([])

# 调整图表的布局
plt.tight_layout()

plt.show()



# lima_compare_vicuna_bench_1k_minipage
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
categories = ['Ours (M)-1k', 'Ours (G)-1k', 'Ours (L)-1k']  # 调整顺序

win_rate_llama = np.array([36.3, 40.6, 35.0])
tie_rate_llama = np.array([28.8, 28.1, 24.4])
loss_rate_llama = np.array([35.0, 31.3, 40.6])

win_rate_mistral = np.array([31.3, 32.5, 26.3])
tie_rate_mistral = np.array([36.9, 40.0, 38.8])
loss_rate_mistral = np.array([31.9, 27.5, 35.0])

# 设置条形宽度
bar_width = 0.2  # 缩小条形的宽度
y_positions = np.arange(len(categories))  # 设置 y 轴位置

# 缩小条形之间的间距
y_spacing = 0.4  # 将每个条形的间隔设置得更紧凑
y_positions = y_positions * y_spacing

# 调整图表的宽度
fig, axs = plt.subplots(1, 2, figsize=(12, 4), gridspec_kw={'wspace': 0})

# 子图1: LLaMA-3.1-8B Base Model
axs[0].barh(y_positions, win_rate_llama, color='#2D9F39', label='Win Rate', height=bar_width, alpha=0.8)
axs[0].barh(y_positions, tie_rate_llama, left=win_rate_llama, color='#F8C03F', label='Tie Rate', height=bar_width, alpha=0.8)
axs[0].barh(y_positions, loss_rate_llama, left=win_rate_llama + tie_rate_llama, color='#E87C2F', label='Loss Rate', height=bar_width, alpha=0.8)

for i in range(len(categories)):
    axs[0].text(win_rate_llama[i] / 2, y_positions[i], f'{win_rate_llama[i]}%', va='center', ha='center', color='black', fontsize=12)
    axs[0].text(win_rate_llama[i] + tie_rate_llama[i] / 2, y_positions[i], f'{tie_rate_llama[i]}%', va='center', ha='center', color='black', fontsize=12)
    axs[0].text(win_rate_llama[i] + tie_rate_llama[i] + loss_rate_llama[i] / 2, y_positions[i], f'{loss_rate_llama[i]}%', va='center', ha='center', color='black', fontsize=12)

# 设置 y 轴的刻度标签和位置
axs[0].set_yticks(y_positions)
axs[0].set_yticklabels([])

# 只保留左侧的竖线
axs[0].spines['left'].set_edgecolor('gray')  # 设置左侧边框颜色为黑色
axs[0].spines['left'].set_linewidth(1.0)          # 调整左侧边框线的宽度
axs[0].spines['top'].set_visible(False)           # 隐藏顶部边框
axs[0].spines['right'].set_visible(False)         # 隐藏右侧边框
axs[0].spines['bottom'].set_visible(False)        # 隐藏底部边框

# 图表设置
axs[0].set_title('Base model: LLaMA-3.1-8B', pad=28, fontsize=22)
axs[0].set_xticks([])

# 子图2: Mistral-7B-v0.3 Base Model
axs[1].barh(y_positions, win_rate_mistral, color='#2D9F39', label='Win Rate', height=bar_width, alpha=0.8)
axs[1].barh(y_positions, tie_rate_mistral, left=win_rate_mistral, color='#F8C03F', label='Tie Rate', height=bar_width, alpha=0.8)
axs[1].barh(y_positions, loss_rate_mistral, left=win_rate_mistral + tie_rate_mistral, color='#E87C2F', label='Loss Rate', height=bar_width, alpha=0.8)

for i in range(len(categories)):
    axs[1].text(win_rate_mistral[i] / 2, y_positions[i], f'{win_rate_mistral[i]}%', va='center', ha='center', color='black', fontsize=12)
    axs[1].text(win_rate_mistral[i] + tie_rate_mistral[i] / 2, y_positions[i], f'{tie_rate_mistral[i]}%', va='center', ha='center', color='black', fontsize=12)
    axs[1].text(win_rate_mistral[i] + tie_rate_mistral[i] + loss_rate_mistral[i] / 2, y_positions[i], f'{loss_rate_mistral[i]}%', va='center', ha='center', color='black', fontsize=12)

# 设置 y 轴的刻度标签和位置
axs[1].set_yticks(y_positions)
axs[1].set_yticklabels([])

# 只保留左侧的竖线
axs[1].spines['left'].set_edgecolor('gray')  # 设置左侧边框颜色为黑色
axs[1].spines['left'].set_linewidth(1.0)          # 调整左侧边框线的宽度
axs[1].spines['top'].set_visible(False)           # 隐藏顶部边框
axs[1].spines['right'].set_visible(False)         # 隐藏右侧边框
axs[1].spines['bottom'].set_visible(False)        # 隐藏底部边框

# 图表设置
axs[1].set_title('Base model: Mistral-7B-v0.3', pad=28, fontsize=22)
axs[1].set_xticks([])

# 合并 legend
axs[1].legend(['Win Rate', 'Tie Rate', 'Loss Rate'], loc='upper center', bbox_to_anchor=(0, 1.15), ncol=3, frameon=False, fontsize=20)

# 调整图表的布局
plt.tight_layout()

plt.show()
```
